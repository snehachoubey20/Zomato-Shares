```python
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
plt.figure(figsize=(10,8))
plt.plot(df.index, df['Close'], marker = 'o', linestyle = '-')
plt.title('Closing Prices over time')
plt.xlabel('Date')
plt.ylabel('Closing Price')
plt.grid(True)
plt.show()
```

```python
plt.figure(figsize=(10,8))
plt.bar(df.index, df['Volume'],width = 1 )
plt.title('Trading volume over time')
plt.xlabel('Date')
plt.ylabel('Volume')
plt.grid(True)
plt.show()
```

```python
plt.figure(figsize=(10,8))
plt.plot(df.index, df['High'], marker = 'o', linestyle = '-')
plt.plot(df.index, df['Low'], marker = 'o', linestyle = '-')
plt.title('High and Low Prices over time')
plt.xlabel('Date')
plt.ylabel('Price')
plt.grid(True)
plt.show()
```

```python
df['Daily Return'] = df['Adj Close'].pct_change()
print(df[['Daily Return']])
```

```python
plt.figure(figsize=(10,8))
plt.bar(df.index, df['Daily Return'],width = 1 )
plt.title('Daily Retruns over time')
plt.xlabel('Date')
plt.ylabel('Daily Returns')
plt.grid(True)
plt.show()
```

```python
df['30 Day MA'] = df['Adj Close'].rolling(window= 30).mean()
df['50 Day MA']= df['Adj Close'].rolling(window= 50).mean()
```

```python
plt.figure(figsize=(10, 5))
plt.plot(df.index, df['Adj Close'], label='Adj Close')
plt.plot(df.index, df['30 Day MA'], label='30 Day MA')
plt.plot(df.index, df['50 Day MA'], label='50 Day MA')
plt.title('Moving Averages')
plt.xlabel('Date')
plt.ylabel('Price')
plt.legend()
plt.grid(True)
plt.show()
```

```python
df['30 Day Std Dev'] = df['Daily Return'].rolling(window=30).std()
```

```python
plt.figure(figsize=(10, 5))
plt.plot(df.index, df['30 Day Std Dev'], label='30 Day Std Dev')
plt.title('Rolling 30 Day Standard Deviation')
plt.xlabel('Date')
plt.ylabel('Standard Deviation')
plt.legend()
plt.grid(True)
plt.show()
```

```python
correlation_matrix = df.corr()
print(correlation_matrix)
```

```python
import matplotlib.pyplot as plt
import mplfinance as mpf
```

```python
def calculate_RSI(data, window=14):
    delta = df['Adj Close'].diff()
    gain = (delta.where(delta > 0, 0)).fillna(0)
    loss = (-delta.where(delta < 0, 0)).fillna(0)
    avg_gain = gain.rolling(window=window).mean()
    avg_loss = loss.rolling(window=window).mean()
    rs = avg_gain / avg_loss
    rsi = 100 - (100 / (1 + rs))
    return rsi
```

```python
df['RSI'] = calculate_RSI(df)


plt.figure(figsize=(10, 5))
plt.plot(df.index, df['RSI'], label='RSI')
plt.axhline(70, color='r', linestyle='--', label='Overbought')
plt.axhline(30, color='g', linestyle='--', label='Oversold')
plt.title('Relative Strength Index (RSI)')
plt.xlabel('Date')
plt.ylabel('RSI')
plt.legend()
plt.grid(True)
plt.show()
```

```python
def calculate_bollinger_bands(data, window=20, num_std=2):
    rolling_mean = df['Adj Close'].rolling(window).mean()
    rolling_std = data['Adj Close'].rolling(window).std()
    data['Bollinger Middle Band'] = rolling_mean
    data['Bollinger Upper Band'] = rolling_mean + (rolling_std * num_std)
    data['Bollinger Lower Band'] = rolling_mean - (rolling_std * num_std)
```

```python
calculate_bollinger_bands(df)


plt.figure(figsize=(10, 5))
plt.plot(df.index, df['Adj Close'], label='Adj Close')
plt.plot(df.index, df['Bollinger Middle Band'], label='Middle Band', linestyle='--')
plt.plot(df.index, df['Bollinger Upper Band'], label='Upper Band', linestyle='--')
plt.plot(df.index, df['Bollinger Lower Band'], label='Lower Band', linestyle='--')
plt.fill_between(df.index, df['Bollinger Upper Band'], df['Bollinger Lower Band'], color='grey', alpha=0.3)
plt.title('Bollinger Bands')
plt.xlabel('Date')
plt.ylabel('Price')
plt.legend()
plt.grid(True)
plt.show()
```
